```python
import matplotlib.pyplot as plt
import librosa as lr
import numpy as np
import random
```

```python
genres = ['disco', 'country', 'jazz', 'classical', 'blues', 'hiphop', 'metal', 'rock', 'pop', 'reggae']
```

```python
audiodata = []

for genre in genres:
    audiodata += [
        lr.load('./dataset/{}/{}.{:05}.wav'.format(str(genre), str(genre), random.randint(0, 99))
    )]
```

```python
y = []
for i in range(10):
    y += [audiodata[i][0]]
```

```python
fig, axs = plt.subplots(10, figsize=[20, 15], sharex=False, sharey=False)

for i, ax in enumerate(axs.flat):
    ax.set_title(genres[i])
    ax.plot(np.linspace(0, y[i].shape[0] / 22050, num=y[i].shape[0]), y[i])
    ax.set_title(
        genres[i],
        fontdict={
            'fontsize': 20
        }
    )
    ax.set_axis_off()

fig.tight_layout(h_pad=2.0)

plt.savefig(
    fname='Tracks.png',
    dpi=400
)

plt.show()
```
