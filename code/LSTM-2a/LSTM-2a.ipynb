```python
from scipy.io.wavfile import read as read_wav
from IPython.display import Audio
import matplotlib.pyplot as plt
import tensorflow as tf
import sklearn as skl
import librosa as lr
import numpy as np
import logging
import random
import scipy
import os
```

### Hyper Parameters:

```python
hparams = {
    'samplerate': 22050,
    'seq_length': 256,
    'hop_length': 512,
    'fft_window': 2048,
    'num_classes': 2
}
```

### Model Creation:

```python
class GenreClassifierModel(tf.keras.Model):
    
    def __init__(self, l1, l2):
        
        super().__init__()
        
        self.L1 = tf.keras.layers.LSTM(l1, return_sequences=True)
        self.L2 = tf.keras.layers.LSTM(l2, return_sequences=False)
        self.L3 = tf.keras.layers.Dense(units=hparams['num_classes'], activation="softmax")
        
    def call(self, inputs):
        self.x = self.L1(inputs)
        self.x = self.L2(self.x)
        self.x = self.L3(self.x)
        return self.x
```

```python
def create_model(config):
    
    l1 = config['L1']
    l2 = config['L2']
    
    # Model Definition:
    model = GenreClassifierModel(l1, l2)
    
    # Model Compilation:
    model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
processed = np.load('../prepared/processed.npy', allow_pickle=True)
```

```python
hist = []
for k in range(0, 6):
    
    train_x = []
    train_y = []

    val_x = []
    val_y = []

    test_x = []
    test_y = []
    
    test_lim_hiphop = 0
    test_lim_metal = 0
    test_lim_rock = 0
    val_lim_hiphop = 0
    val_lim_metal = 0
    val_lim_rock = 0
    train_lim_hiphop = 0
    train_lim_metal = 0
    train_lim_rock = 0
    
    for entry in processed:
        # Test Set:
        if entry['index'] >= 90 and entry['index'] < 100:
            if (entry['ground_truth'] == 'hiphop') and test_lim_hiphop < 33:
                test_x += [entry['features']]
                test_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                test_lim_hiphop += 1
            if (entry['ground_truth'] == 'metal') and test_lim_metal < 34:
                test_x += [entry['features']]
                test_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                test_lim_metal += 1
            if (entry['ground_truth'] == 'rock') and test_lim_rock < 33:
                test_x += [entry['features']]
                test_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                test_lim_rock += 1
            if entry['ground_truth'] in ['pop', 'reggae']:
                test_x += [entry['features']]
                test_y += [tf.keras.utils.to_categorical(1, num_classes=hparams['num_classes'])]
                
        # Validation Set:
        elif entry['index'] >= (k * 15) and entry['index'] < ((k+1) * 15):
            if entry['ground_truth'] == 'hiphop' and val_lim_hiphop < 50:
                val_x += [entry['features'].copy()]
                val_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                val_lim_hiphop += 1
            if entry['ground_truth'] == 'metal' and val_lim_metal < 50:
                val_x += [entry['features'].copy()]
                val_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                val_lim_metal += 1
            if entry['ground_truth'] == 'rock' and val_lim_rock < 50:
                val_x += [entry['features'].copy()]
                val_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                val_lim_rock += 1
            if entry['ground_truth'] in ['pop', 'reggae']:
                val_x += [entry['features'].copy()]
                val_y += [tf.keras.utils.to_categorical(1, num_classes=hparams['num_classes'])]
        
        # Train Set:
        else:
            if entry['ground_truth'] == 'hiphop' and train_lim_hiphop < 250:
                train_x += [entry['features'].copy()]
                train_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                train_lim_hiphop += 1
            if entry['ground_truth'] == 'metal' and train_lim_metal < 250:
                train_x += [entry['features'].copy()]
                train_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                train_lim_metal += 1
            if entry['ground_truth'] == 'rock' and train_lim_rock < 250:
                train_x += [entry['features'].copy()]
                train_y += [tf.keras.utils.to_categorical(0, num_classes=hparams['num_classes'])]
                train_lim_rock += 1
            if entry['ground_truth'] in ['pop', 'reggae']:
                train_x += [entry['features'].copy()]
                train_y += [tf.keras.utils.to_categorical(1, num_classes=hparams['num_classes'])]
    
    train_x = np.array(train_x)
    train_y = np.array(train_y)

    val_x = np.array(val_x)
    val_y = np.array(val_y)

    test_x = np.array(test_x)
    test_y = np.array(test_y)
    
    # Experiment:
    model = create_model({
        'L1': 64,
        'L2': 32
    })
    
    
    with open('k-fold-train-loss.txt', 'a') as file:
        print('===========================================================', file=file)
        print('Fold Number:', k, file=file)
    
    hist += [model.fit(train_x, train_y, epochs=100, verbose=0)]
    
    with open('k-fold-train-loss.txt', 'a') as file:
        
        print('Train Accuracy: {}'.format(
            100 * model.evaluate(train_x, train_y, verbose=0)[1]
        ), file=file)
        
        print('Val Accuracy: {}'.format(
            100 * model.evaluate(val_x, val_y, verbose=0)[1]
        ), file=file)
```
